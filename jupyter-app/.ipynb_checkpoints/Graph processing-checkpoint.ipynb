```python
# Importing
import os
import json
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from matplotlib.colors import TABLEAU_COLORS 
```

```python
# Defining functions
def calculate_magnitude(accel):
    return np.sqrt(accel['x']**2 + accel['y']**2 + accel['z']**2)

def process_file(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    
    timestamps = []
    magnitudes = []
    
    for point in data:
        timestamps.append(point['timestamp'] / 1000)
        magnitudes.append(calculate_magnitude(point['acceleration']))
    
    return timestamps, magnitudes
```

```python
file_pattern = '*.json'
files = sorted(glob("./data/" + file_pattern))
y_min, y_max = 0.6, 1.7
```

```python
max_individual_plots = 18
min_individual_plots = max_individual_plots - 6

if not files:
    print("No JSON files found in the current directory!")
else:
    print(f"Found {len(files)} JSON files to process")
if not files:
    print("No JSON files found in the specified folder!")

else:
    print(f"Found {len(files)} JSON files. Generating plots...")

    for file_path in files:
        timestamps, magnitudes = process_file(file_path)
        
    print("\n=== Individual Plots ===")
    for i, file_path in enumerate(files[min_individual_plots:max_individual_plots]):
        timestamps, magnitudes = process_file(file_path)
        if timestamps:
            plt.figure(figsize=(10, 4))
            plt.plot(np.array(timestamps)-timestamps[0], magnitudes, 'b-')
            plt.ylim(y_min, y_max)
            plt.title(f'Acceleration: {os.path.basename(file_path)}')
            plt.xlabel('Time (s)'); plt.ylabel('Magnitude (m/s)')
            plt.grid(True)
            plt.tight_layout()
            plt.show()
```

```python
if not files:
    print("No JSON files found in the current directory!")
else:
    print(f"Found {len(files)} JSON files to process")
if not files:
    print("No JSON files found in the specified folder!")

else:
    print(f"Found {len(files)} JSON files. Generating plots...")

    print("\n=== Combined Plot ===")
    plt.figure(figsize=(12, 6))
    colors = list(TABLEAU_COLORS.values())
    all_magnitudes = []
    
    for i, file_path in enumerate(files[min_individual_plots:max_individual_plots]):
        timestamps, magnitudes = process_file(file_path)
        if timestamps:
            time_normalized = np.array(timestamps) - timestamps[0]
            line = plt.plot(time_normalized, magnitudes, 
                          color=colors[i % len(colors)],
                          label=os.path.basename(file_path),
                          alpha=0.5)
            all_magnitudes.append(magnitudes)
    
    if all_magnitudes:
        min_length = min(len(mag) for mag in all_magnitudes)
        trimmed_mags = [mag[:min_length] for mag in all_magnitudes]
        median_magnitude = np.median(trimmed_mags, axis=0)
        median_timestamps = time_normalized[:min_length]
        
        plt.plot(median_timestamps, median_magnitude, 
                'k-', linewidth=3, label='Median', alpha=0.9)
    
    plt.ylim(y_min, y_max)
    plt.title('Acceleration Comparison with Average')
    plt.xlabel('Time (s)'); plt.ylabel('Magnitude (m/s)')
    plt.grid(True)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
    def get_median_data(file_slice):
        """Helper to get median data for a file slice"""
        magnitudes_list = []
        for file_path in file_slice:
            timestamps, magnitudes = process_file(file_path)
            if timestamps:
                time_normalized = np.array(timestamps) - timestamps[0]
                magnitudes_list.append(magnitudes)
        if magnitudes_list:
            min_length = min(len(mag) for mag in magnitudes_list)
            trimmed = [mag[:min_length] for mag in magnitudes_list]
            return time_normalized[:min_length], np.median(trimmed, axis=0)
        return None, None

    # Compare bike and car
    if len(files) >= 12:
        print("\n=== Median Comparison: Bike vs Car at 30 km/h ===")
        plt.figure(figsize=(12, 6))
        
        x1, y1 = get_median_data(files[:6])
        if x1 is not None:
            plt.plot(x1, y1, 'b-', linewidth=2, label='Median Bike 30 km/h', alpha=0.8)
        
        x2, y2 = get_median_data(files[6:12])
        if x2 is not None:
            plt.plot(x2, y2, 'r-', linewidth=2, label='Median Car 30 km/h', alpha=0.8)
        
        # plt.ylim(y_min, y_max)
        plt.title('Median Comparison: Car vs Bike 30 km/h')
        plt.xlabel('Time (s)'); plt.ylabel('Magnitude (m/s)')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

    # Compare different speeds
    if len(files) >= 18:
        print("\n=== Median Comparison: Batches 7-12 vs 13-18 ===")
        plt.figure(figsize=(12, 6))
        
        if x2 is not None:
            plt.plot(x2, y2, 'r-', linewidth=2, label='Median Car 30 km/h', alpha=0.8)
        

        x3, y3 = get_median_data(files[12:18])
        if x3 is not None:
            plt.plot(x3, y3, 'g-', linewidth=2, label='Median Car 40 km/h', alpha=0.8)
        
        # plt.ylim(y_min, y_max)
        plt.title('Median Comparison: Car at 30 km/h vs 40 km/h')
        plt.xlabel('Time (s)'); plt.ylabel('Magnitude (m/s)')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()
```

```python

def get_extreme_data(file_slice):
    """Helper to get extreme values (farthest from 1) for a file slice"""
    magnitudes_list = []
    for file_path in file_slice:
        timestamps, magnitudes = process_file(file_path)
        if timestamps:
            time_normalized = np.array(timestamps) - timestamps[0]
            magnitudes_list.append(magnitudes)
    if magnitudes_list:
        min_length = min(len(mag) for mag in magnitudes_list)
        trimmed = [mag[:min_length] for mag in magnitudes_list]
        
        extremes = []
        for i in range(min_length):
            values_at_point = [mag[i] for mag in trimmed]
            # Get value farthest from 1 (using absolute difference)
            extreme_val = max(values_at_point, key=lambda x: abs(x - 1))
            extremes.append(extreme_val)
            
        return time_normalized[:min_length], np.array(extremes)
    return None, None

# Compare extremes: bike vs car
if len(files) >= 12:
    print("\n=== Extreme Values Comparison: Batches Bike vs Car ===")
    plt.figure(figsize=(12, 6))
    
    x1, y1 = get_extreme_data(files[:6])
    if x1 is not None:
        plt.plot(x1, y1, 'b-', linewidth=2, label='Extremes 1-6', alpha=0.8)
    
    x2, y2 = get_extreme_data(files[6:12])
    if x2 is not None:
        plt.plot(x2, y2, 'r-', linewidth=2, label='Extremes 7-12', alpha=0.8)
    
    plt.axhline(y=1, color='k', linestyle='--', alpha=0.3)  # Reference line at 1
    plt.title('Extreme Values Comparison: Bike vs Car')
    plt.xlabel('Time (s)'); plt.ylabel('Magnitude (m/s)')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

if len(files) >= 18:
    print("\n=== Extreme Values Comparison: Batches Car at 30 km/h vs 40 km/h ===")
    plt.figure(figsize=(12, 6))
    
    if x2 is not None:
        plt.plot(x2, y2, 'r-', linewidth=2, label='Extremes 7-12', alpha=0.8)
    
    x3, y3 = get_extreme_data(files[12:18])
    if x3 is not None:
        plt.plot(x3, y3, 'g-', linewidth=2, label='Extremes 13-18', alpha=0.8)
    
    plt.axhline(y=1, color='k', linestyle='--', alpha=0.3)
    plt.title('Extreme Values Comparison: ar at 30 km/h vs 40 km/h')
    plt.xlabel('Time (s)'); plt.ylabel('Magnitude (m/s)')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
